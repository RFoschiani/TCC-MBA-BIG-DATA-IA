```python
# prompt: a partir do resultado anterior, Faça um crawler que salve em um dataframe esse link e todos os links dessa página no dataframe. Imprima-os também,  Precisa desconsiderar os link que contenham "Pesq=&anoPesq="

!pip install beautifulsoup4 requests

import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawl_and_extract_links(url):
    """
    Crawls a given URL, extracts all links on the page, filters them,
    and returns a list of filtered links.
    """
    try:
        response = requests.get(url)
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
        soup = BeautifulSoup(response.content, 'html.parser')
        links = [a['href'] for a in soup.find_all('a', href=True)]
        filtered_links = [link for link in links if "Pesq=&anoPesq=" not in link]
        return filtered_links
    except requests.exceptions.RequestException as e:
        print(f"Error crawling {url}: {e}")
        return []

# Create an empty list to store all extracted links
all_links = []

# Assuming `urls` contains the list of URLs generated in the previous code block
# Iterate through the generated URLs and crawl each one
for url in urls:
    extracted_links = crawl_and_extract_links(url)
    all_links.extend(extracted_links)
    # Also include the current URL in the list of links
    if "Pesq=&anoPesq=" not in url:
      all_links.append(url)


# Create a pandas DataFrame from the list of all links
df_links = pd.DataFrame({'Link': all_links})

# Print the DataFrame and the links
df_links

# Or print the links directly from the list
# for link in all_links:
#     print(link)
```

```python
# prompt: Usando o DataFrame df_links: para cada linha, utilize esse link e faça o download de todo o link que leve a um arquivo pdf, e salve na pasta do google drive MBABIGDATAIA e subpsta DATASET, ao final imprim o total de PDF que foi feito o download

import requests
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Define a pasta de destino no Google Drive
output_folder = '/content/drive/MyDrive/MBABIGDATAIA/DATASET'

# Cria a pasta de destino se ela não existir
os.makedirs(output_folder, exist_ok=True)

# Inicializa um contador para PDFs baixados e uma lista para erros
downloaded_pdf_count = 0
error_links = []

# Itera através de cada linha do DataFrame
for index, row in df_links.iterrows():
    link = row['Link']
    try:
        # Busca o conteúdo do link
        response = requests.get(link)
        response.raise_for_status()  # Lança uma exceção para códigos de status ruins

        # Analisa o conteúdo HTML
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontra todos os links que apontam para arquivos PDF
        pdf_links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].lower().endswith('.pdf')]

        # Se o link original já for um PDF, adiciona-o à lista
        if link.lower().endswith('.pdf'):
            pdf_links.append(link)

        # Baixa cada link de PDF encontrado
        for pdf_link in pdf_links:
            # Garante que o link seja absoluto
            if not pdf_link.startswith('http'):
                pdf_link = urljoin(link, pdf_link)

            try:
                # Extrai o nome do arquivo da URL
                filename = os.path.basename(pdf_link)
                file_path = os.path.join(output_folder, filename)

                # NOVO: Verifica se o arquivo já existe na pasta de destino
                if os.path.exists(file_path):
                    print(f"PULADO: O arquivo {filename} já existe.")
                    continue  # Pula para o próximo pdf_link

                # Baixa o PDF
                pdf_response = requests.get(pdf_link, stream=True)
                pdf_response.raise_for_status()

                with open(file_path, 'wb') as f:
                    for chunk in pdf_response.iter_content(chunk_size=8192):
                        f.write(chunk)

                downloaded_pdf_count += 1
                print(f"BAIXADO: {filename}")

            except requests.exceptions.RequestException as e:
                print(f"ERRO AO BAIXAR o PDF {pdf_link}: {e}")
                error_links.append(pdf_link)  # Adiciona o link do PDF com erro à lista

    except requests.exceptions.RequestException as e:
        print(f"ERRO AO ACESSAR o link {link}: {e}")
        error_links.append(link)  # Adiciona o link principal com erro à lista
    except Exception as e:
        print(f"Um erro inesperado ocorreu para o link {link}: {e}")
        error_links.append(link)  # Adiciona o link principal com erro à lista


# NOVO: Salva todos os links que tiveram erro em um arquivo
if error_links:
    error_file_path = os.path.join(output_folder, 'DOWNLOAD_ERROR.txt')
    with open(error_file_path, 'w') as f:
        f.write("DOWNLOAD_ERROR:\n")
        for error_link in error_links:
            f.write(f"{error_link}\n")
    print(f"\nLog de erros salvo em: {error_file_path}")


# Imprime o número total de PDFs baixados
print(f"\nTotal de PDFs baixados com sucesso: {downloaded_pdf_count}")
```

```python
import requests
import os

# Define the path to the error file and the output folder
error_file_path = '/content/drive/MyDrive/MBABIGDATAIA/DATASET/DOWNLOAD_ERROR.txt'
output_folder = '/content/drive/MyDrive/MBABIGDATAIA/DATASET'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Initialize a counter for downloaded PDFs
downloaded_cedoc_count = 0

# Read the error file and filter the links
cedoc_links = []
try:
    with open(error_file_path, 'r') as f:
        for line in f:
            line = line.strip()
            # Clean the URL: remove everything before "www" and reconstruct
            if "www" in line:
                line = "https://" + line.split("www.", 1)[-1]

            if "https://www2.aneel.gov.br/cedoc/" in line and line.lower().endswith('.pdf'):
                cedoc_links.append(line)
except FileNotFoundError:
    print(f"Error: The file {error_file_path} was not found.")

# Download the filtered PDF links
print(f"Attempting to download {len(cedoc_links)} PDFs from cedoc...")
for pdf_link in cedoc_links:
    try:
        # Extract the filename from the URL
        filename = os.path.basename(pdf_link)
        file_path = os.path.join(output_folder, filename)

        # Check if the file already exists
        if os.path.exists(file_path):
            print(f"PULADO: O arquivo {filename} já existe.")
            continue

        # Download the PDF
        pdf_response = requests.get(pdf_link, stream=True)
        pdf_response.raise_for_status()

        with open(file_path, 'wb') as f:
            for chunk in pdf_response.iter_content(chunk_size=8192):
                f.write(chunk)

        downloaded_cedoc_count += 1
        print(f"BAIXADO: {filename}")

    except requests.exceptions.RequestException as e:
        print(f"ERRO AO BAIXAR o PDF {pdf_link}: {e}")
    except Exception as e:
        print(f"Um erro inesperado ocorreu para o link {pdf_link}: {e}")


# Print the total number of PDFs downloaded
print(f"\nTotal de PDFs baixados com sucesso da pasta cedoc: {downloaded_cedoc_count}")
```
