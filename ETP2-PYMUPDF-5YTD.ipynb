```python
# É necessário instalar a biblioteca PyMuPDF
!pip install pymupdf

import os
import pymupdf  # PyMuPDF
from google.colab import drive

# Monta o Google Drive para que o Colab possa acessar os arquivos
drive.mount('/content/drive')

# Caminho para a pasta contendo os arquivos PDF
pasta_alvo = '/content/drive/MyDrive/MBABIGDATAIA/DATASET'

# Verifica se o diretório existe
if not os.path.isdir(pasta_alvo):
    print(f"Erro: O diretório '{pasta_alvo}' não foi encontrado.")
else:
    # Itera sobre todos os arquivos na pasta
    for nome_arquivo in os.listdir(pasta_alvo):
        # Verifica se o arquivo é um PDF
        if nome_arquivo.endswith(".pdf"):
            # Constrói o caminho completo para o arquivo PDF
            caminho_pdf = os.path.join(pasta_alvo, nome_arquivo)

            # Cria o nome do arquivo de saída .txt
            nome_arquivo_saida = os.path.splitext(nome_arquivo)[0] + ".txt"
            caminho_txt = os.path.join(pasta_alvo, nome_arquivo_saida)

            # Verifica se o arquivo .txt já existe
            if os.path.exists(caminho_txt):
                print(f"Arquivo '{nome_arquivo_saida}' já existe. Pulando a conversão de '{nome_arquivo}'.")
                continue  # Pula para o próximo arquivo PDF

            # Abre o arquivo PDF
            doc = pymupdf.open(caminho_pdf)
            texto_completo = ''

            # Extrai o texto de cada página do PDF
            for page in doc:
                texto_completo += page.get_text()

            # Fecha o documento PDF
            doc.close()

            # Salva o texto extraído no arquivo .txt
            with open(caminho_txt, 'w', encoding='utf-8') as arquivo_saida:
                arquivo_saida.write(texto_completo)

            print(f"Texto extraído de '{nome_arquivo}' e salvo em '{nome_arquivo_saida}'")

print("\nProcesso concluído.")
```
